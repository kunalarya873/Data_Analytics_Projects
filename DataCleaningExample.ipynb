**handling the Problem of Na, N/a**

```python
missing_value=["N/a","na",np.nan]
df = pd.read_csv("tempData.csv", na_values=missing_value)
```

```python
# checking again
df.isnull().sum()
```

```python
# another method to check if any NA values
df.isnull().any()
```

```python
# visualizing the NA values
sns.heatmap(df.isnull(), yticklabels=False)
```

```python
# visualizing the NA values
sns.heatmap(df.isnull(), yticklabels=False, annot=True)
```
